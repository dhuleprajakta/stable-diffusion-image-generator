```python
from diffusers import StableDiffusionPipeline
import torch

pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
pipe = pipe.to("cuda")

image = pipe("A capybara holding a sign that reads Hello World").images[0]
image.save("capybara.png")
```

```python
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image
import matplotlib.pyplot as plt

# Prompt the user for input text
text_input = input("Enter the description for the image: ")

# Check if CUDA is available before moving the pipeline
if torch.cuda.is_available():
    pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
    pipe = pipe.to("cuda")

    image = pipe(text_input).images[0]
    image.save("generated_image.png")
else:
    print("CUDA is not available. Running on CPU instead.")
    # If CUDA is not available, load the pipeline on CPU
    pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5")
    image = pipe(text_input).images[0]
    image.save("generated_image.png")

# Display the image using matplotlib
plt.imshow(image)
plt.axis('off')  # Turn off axis
plt.show()

```
